```python
#Converting the Weight column to kg and also change to int
def fun(x):
    if 'lbs' in x:
        y = x.replace('lbs', '')
        y = int(y) * 0.454
        return int(round(y))
    else:
        return int(x.replace('kg', ''))
    
    
df['Weight'] = df['Weight'].map(fun)
```

```python
df['Weight'].unique()
```

```python
# Convert the Height column to cm from inches and also to intergers
def func(x):
    if '"' in x:
        y = x.replace('"', '')
        y = y.split("\'")
        y = int(y[0]) + int(y[1]) * 0.0833
        y = 30.48 * y
        return int(round(y))
    else:
        return int(x.replace('cm', ''))
    
df['Height'] = df['Height'].map(func)
```
